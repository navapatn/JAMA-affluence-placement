```python
import pandas as pd

file_path = '/content/drive/My Drive/JAMA-ADI-data/DAC_NationalDownloadableFile.csv'
df = pd.read_csv(file_path)

selected_columns = [
    'NPI', 'gndr', 'Med_sch', 'Grd_yr',  'pri_spec', 'ZIP Code', 'City/Town', 'State'
]

df_filtered = df[selected_columns]
df_filtered.head()


```

```python
import os
import pandas as pd

adi_2015_dir = '/content/drive/My Drive/JAMA-ADI-data/2015/'

def load_and_merge_adi_files(directory):
    merged_data = pd.DataFrame()  # Initialize empty DataFrame
    for filename in os.listdir(directory):
        if filename.endswith('.txt'):
            state = filename.split('_')[0]
            file_path = os.path.join(directory, filename)

            adi_data = pd.read_csv(file_path)


            adi_data['state'] = state
            merged_data = pd.concat([merged_data, adi_data], ignore_index=True)

    return merged_data

adi_2015_data = load_and_merge_adi_files(adi_2015_dir)

adi_2015_data.head()
```

```python
import os
import pandas as pd

adi_2020_dir = '/content/drive/My Drive/JAMA-ADI-data/2020/'


def load_and_merge_adi_2020_files(directory):
    merged_data = pd.DataFrame()
    for filename in os.listdir(directory):
        if filename.endswith('.csv'):
            state = filename.split('_')[0]
            file_path = os.path.join(directory, filename)

            adi_data = pd.read_csv(file_path)

            adi_data['state'] = state
            merged_data = pd.concat([merged_data, adi_data], ignore_index=True)

    return merged_data

adi_2020_data = load_and_merge_adi_2020_files(adi_2020_dir)
adi_2020_data.head()
```

```python

adi_2020_data['ZIP_Code_5_Digit_Standardized'] = adi_2020_data['ZIPID'].astype(str).str[1:6]

df_filtered['ZIP_Code_5_Digit_Standardized'] = df_filtered['ZIP Code'].astype(str).str[:5]

zip_comparison = pd.DataFrame({
    'Medical Dataset ZIP (Standardized)': df_filtered['ZIP_Code_5_Digit_Standardized'].head(),
    'ADI Dataset ZIP (Standardized)': adi_2020_data['ZIP_Code_5_Digit_Standardized'].head()
})

zip_comparison
```

```python
# Step 1: Extract unique ZIP codes from the medical dataset where grad_year >= 2015
unique_medical_zips_recent = set(df_filtered[df_filtered['Grd_yr'] >= 2015]['ZIP_Code_5_Digit_Standardized'].unique())

# Step 2: Extract unique ZIP codes from the ADI dataset
unique_adi_zips = set(adi_2020_data['ZIP_Code_5_Digit_Standardized'].unique())

# Step 3: Compare the two sets to find common ZIP codes
common_zips = unique_medical_zips_recent.intersection(unique_adi_zips)

# Step 4: Count the number of unique ZIP codes in each dataset and the common ones
num_unique_medical_zips_recent = len(unique_medical_zips_recent)
num_unique_adi_zips = len(unique_adi_zips)
num_common_zips = len(common_zips)

# Display the results
print(f"Number of unique ZIP codes in the filtered medical dataset (grad_year >= 2015): {num_unique_medical_zips_recent}")
print(f"Number of unique ZIP codes in the ADI dataset: {num_unique_adi_zips}")
print(f"Number of common ZIP codes between the two datasets: {num_common_zips}")

# Calculate the percentage of medical ZIP codes that matched with ADI dataset
percentage_match = (num_common_zips / num_unique_medical_zips_recent) * 100
print(f"Percentage of medical ZIP codes (grad_year >= 2015) that matched with ADI dataset: {percentage_match:.2f}%")
```

```python
import pandas as pd

adi_2020_data['ZIP_Code_5_Digit_Standardized'] = adi_2020_data['BENE_ZIP_CD'].astype(str).str[:5]


adi_2020_data['ADI_NATRANK'] = pd.to_numeric(adi_2020_data['ADI_NATRANK'], errors='coerce')
adi_2020_data['ADI_STATERNK'] = pd.to_numeric(adi_2020_data['ADI_STATERNK'], errors='coerce')

adi_2020_data_clean = adi_2020_data.dropna(subset=['ADI_NATRANK', 'ADI_STATERNK'])

adi_aggregated_2020 = adi_2020_data_clean.groupby('ZIP_Code_5_Digit_Standardized').agg({
    'ADI_NATRANK': 'mean',   # Average the national rank
    'ADI_STATERNK': 'mean',  # Average the state rank
    'GISJOIN': 'first',      # Take the first GISJOIN as an example (or adjust as needed)
    'FIPS': 'first',         # Take the first FIPS code
}).reset_index()


adi_aggregated_2020.rename(columns={
    'ADI_NATRANK': 'Avg_ADI_NATRANK',
    'ADI_STATERNK': 'Avg_ADI_STATERNK'
}, inplace=True)


print(adi_aggregated_2020.head())
```

```python
import pandas as pd


adi_2015_data['ZIP_Code_5_Digit_Standardized'] = adi_2015_data['ZIPID'].str[1:6]

adi_2015_data['ADI_NATRANK'] = pd.to_numeric(adi_2015_data['ADI_NATRANK'], errors='coerce')
adi_2015_data['ADI_STATERNK'] = pd.to_numeric(adi_2015_data['ADI_STATERNK'], errors='coerce')


adi_2015_data_clean = adi_2015_data.dropna(subset=['ADI_NATRANK', 'ADI_STATERNK'])


adi_aggregated_2015 = adi_2015_data_clean.groupby('ZIP_Code_5_Digit_Standardized').agg({
    'ADI_NATRANK': 'mean',   # Average the national rank
    'ADI_STATERNK': 'mean',  # Average the state rank
    'GISJOIN': 'first',      # Take the first GISJOIN as an example (or adjust as needed)
}).reset_index()


adi_aggregated_2015.rename(columns={
    'ADI_NATRANK': 'Avg_ADI_NATRANK',
    'ADI_STATERNK': 'Avg_ADI_STATERNK'
}, inplace=True)

adi_aggregated_2015.head()
```

```python

adi_aggregated_2020['Avg_ADI_NATRANK'].fillna(adi_aggregated_2020['Avg_ADI_NATRANK'].mean(), inplace=True)
adi_aggregated_2020['Avg_ADI_STATERNK'].fillna(adi_aggregated_2020['Avg_ADI_STATERNK'].mean(), inplace=True)

adi_aggregated_2015['Avg_ADI_NATRANK'].fillna(adi_aggregated_2015['Avg_ADI_NATRANK'].mean(), inplace=True)
adi_aggregated_2015['Avg_ADI_STATERNK'].fillna(adi_aggregated_2015['Avg_ADI_STATERNK'].mean(), inplace=True)
```

```python
df_filtered_recent_grads = df_filtered[df_filtered['Grd_yr'] >= 2015]  # Already filtered for grad_year >= 2015


merged_with_2015 = pd.merge(df_filtered_recent_grads, adi_aggregated_2015,
                            left_on='ZIP_Code_5_Digit_Standardized',
                            right_on='ZIP_Code_5_Digit_Standardized',
                            how='left')


merged_with_2015_and_2020 = pd.merge(merged_with_2015, adi_aggregated_2020,  # adi_aggregated is for 2020 ADI
                                     left_on='ZIP_Code_5_Digit_Standardized',
                                     right_on='ZIP_Code_5_Digit_Standardized',
                                     how='left',
                                     suffixes=('_2015', '_2020'))

merged_with_2015_and_2020.head()
```

```python
merged_with_2015_and_2020.to_csv('medical_doctors_with_ADI.csv', index=False)
```
